**2) Question ( Based on Value Counts )**
- Check what are the different types of make are there in out dataset. and what is the count ( occurence) of each Make in the data ?

```python
car['Make'].value_counts()
```

**3) Instruction(Filtering)**
- Show all the Records when origin is Asia or Europe

```python
# 1st Method to obtain all records using 'or' operator

car[(car['Origin']== 'Asia') | (car['Origin']== 'Europe')]
```

```python
# 2nd Method to obtain all records using 'isin' Function

car[car['Origin'].isin(['Asia','Europe'])]
```

**4) Instruction(Removing unwanted records)**
- Remove all the recents (rows) where Weight is above 4000

```python
car[~(car['Weight'] > 4000)]                      # ~ this Sign is remover
```

**5) Instruction (Applying function on a column)**
- Increase all the values of "MPG_City" column by 3

```python
car.head(2)
```

```python
car["MPG_City"] = car["MPG_City"].apply(lambda x:x+3)
```
